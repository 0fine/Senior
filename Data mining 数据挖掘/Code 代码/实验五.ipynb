```python
import numpy as np  # 导入NumPy库，用于科学计算和数组操作，简写为np
import pandas as pd  # 导入Pandas库，用于数据处理和分析，简写为pd
from sklearn import datasets  # 从scikit-learn导入数据集模块，用于加载内置数据集

from rulefit import RuleFit  # 从rulefit库导入RuleFit类，用于构建可解释的规则模型
from sklearn.ensemble import GradientBoostingRegressor  # 从scikit-learn导入梯度提升回归器，用于构建集成回归模型

# 从原始URL加载波士顿数据集
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
# 提取特征和目标值
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
# 特征名称（与原数据集一致）
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
                 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']


y = target   # 提取目标变量（房价中位数）
X = pd.DataFrame(data)  # 将特征数据转换为DataFrame格式
X.columns = feature_names  # 为特征DataFrame设置列名（使用数据集中的特征名），设置特征列名
X['medv'] = y.astype('int')  # 新增'medv'列存储目标变量，并转换为整数类型（原目标为浮点数）
features = feature_names  # 提取特征名称列表，便于后续使用

X1 = X[feature_names]  # 从X中筛选出原始特征列，作为模型的输入特征
y1 = X['medv']  # 从X中提取'medv'列作为模型的目标变量（后续建模用）
```

```python
print(y)
print(X1)
```

```python
import warnings
# 忽略相关警告
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# 初始化RuleFit模型实例，RuleFit是一种结合决策规则和线性模型的可解释性机器学习模型
rf = RuleFit()

# 使用训练数据拟合RuleFit模型
# X1.values：输入的特征数据（转换为NumPy数组格式，RuleFit模型通常接受数组输入）
# y1：对应的目标变量（即需要预测的标签，这里是房价中位数'medv'）
# features：特征名称列表，用于在生成规则时标注特征，提升模型解释性
rf.fit(X1.values, y1, features)
```

```python
# 从训练好的RuleFit模型中提取所有规则（包括决策规则和线性项）
# 规则包含规则文本、系数(coef)、支持度(support)等信息
rules = rf.get_rules()

# 筛选出系数不为0的规则（排除无贡献的规则），并按支持度(support)降序排序
# coef != 0：系数为0的规则对模型预测无影响，可过滤掉
# sort_values("support", ascending=False)：按支持度从高到低排序，支持度表示规则在训练数据中满足的比例
rules = rules[rules.coef != 0].sort_values("support", ascending=False)

# 打印筛选并排序后的规则，便于查看模型中重要的决策逻辑
print(rules)
```
