ArticlesAll对象
一个用于存储跨文章的术语数据的对象。

    属性
    ----------
    label : str
        术语的标签。
    term : Term
        搜索术语的定义，包括包含和排除的单词。
    has_data : bool
        对象是否包含数据。
    n_articles : int
        对象中包含的文章数量。
    ids : int列表
        对象中所有文章的ID列表。
    journals : collections.Counter
        每个期刊的频率分布。
    authors : collections.Counter
        每个作者的频率分布。
    first_authors : collections.Counter
        每个第一作者的频率分布。
    last_authors : collections.Counter
        每个最后作者的频率分布。
    words : collections.Counter
        所有单词的频率分布。
    keywords : collections.Counter
        所有关键词的频率分布。
    years : collections.Counter
        每个出版年份的频率分布。
    dois : str列表
        对象中每篇文章的DOI。
    summary : dict
        与当前对象关联的数据摘要。
